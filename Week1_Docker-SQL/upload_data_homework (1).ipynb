## Question 3

```python
### Question3
filtered_df = df[(df['lpep_pickup_datetime'] >= '2019-10-01') & (df['lpep_dropoff_datetime'] < '2019-11-01')]

count_up_to_1_mile = len(filtered_df[filtered_df['trip_distance'] <= 1])
count_1_to_3_miles = len(filtered_df[(filtered_df['trip_distance'] > 1) & (filtered_df['trip_distance'] <= 3)])
count_3_to_7_miles = len(filtered_df[(filtered_df['trip_distance'] > 3) & (filtered_df['trip_distance'] <= 7)])
count_7_to_10_miles = len(filtered_df[(filtered_df['trip_distance'] > 7) & (filtered_df['trip_distance'] <= 10)])
count_over_10_miles = len(filtered_df[filtered_df['trip_distance'] > 10])

print("Up to 1 mile:", count_up_to_1_mile)
print("1 < distance <= 3 miles:", count_1_to_3_miles)
print("3 < distance <= 7 miles:", count_3_to_7_miles)
print("7 < distance <= 10 miles:", count_7_to_10_miles)
print("Over 10 miles:", count_over_10_miles)
```
